```python
value_vars = list(df_modified.columns)[8:]
value_vars
```

```python
df_melted = df_modified.melt(id_vars=['Respondent ID-', 'Identify which division you work in.-Response',
       'Identify which division you work in.-Other (please specify)',
       'Which of the following best describes your position level?-Response',
       'Which generation are you apart of?-Response',
       'Please select the gender in which you identify.-Response',
       'Which duration range best aligns with your tenure at your company?-Response',
       'Which of the following best describes your employment type?-Response'],value_vars=value_vars,var_name='Question+Subquestion',value_name='Answer')
df_melted
```

```python
df_ques = pd.read_excel(pwd+'\edited survey monkey.xlsx',sheet_name='Question')
df_ques
```

```python
ques = df_ques.copy()
```

```python
ques.columns
```

```python
ques.drop(columns=['raw_question', 'raw_subquestion'],inplace=True)
```

```python
ques.drop(columns=['subquestion'],inplace=True)
```

```python
ques
```

```python
ques.dropna(inplace=True)
```

```python
ques
```

```python
df_merged = pd.merge(left= df_melted, right=ques, how="left", left_on="Question+Subquestion", right_on="Question+Subquestion")
```

```python
df_merged
```

```python
print("Original Data: ",len(df_melted))
print("Final Data: ",len(df_merged))
```

```python
respondnts = df_merged[df_merged["Answer"].notna()]
respondnts= respondnts.groupby("question")["Respondent ID-"].nunique().reset_index()
respondnts.rename(columns={'Respondent ID-':'Respondents'},inplace=True)
respondnts
```

```python
df_merged[df_merged['Answer'].notna()]
```

```python
df_merged_2 = pd.merge(left= df_merged, right=respondnts, how="left", left_on="question", right_on="question")
df_merged_2
```

```python
print(len(df_merged))
print(len(df_merged_2))
```

```python
same_ans = df_merged #[df_merged["Answer"].notna()]
same_ans= same_ans.groupby(['Question+Subquestion','Answer'])["Respondent ID-"].nunique().reset_index()
same_ans.rename(columns={'Respondent ID-':'Same Answer'},inplace=True)
same_ans
```

```python
df_merged_3 = pd.merge(left= df_merged_2, right=same_ans, how="left", left_on=['Question+Subquestion','Answer'], right_on=['Question+Subquestion','Answer'])
df_merged_3['Same Answer'].fillna(0,inplace=True)
df_merged_3
```

```python
df_merged_3.columns
```

```python
output = df_merged_3.copy()
output.rename(columns={'Identify which division you work in.-Response':"Division Primary", 'Identify which division you work in.-Other (please specify)':"Division Secondary", 'Which of the following best describes your position level?-Response':"Position", 'Which generation are you apart of?-Response':"Generation", 'Please select the gender in which you identify.-Response':"Gender", 'Which duration range best aligns with your tenure at your company?-Response':"Tenure", 'Which of the following best describes your employment type?-Response':"Employment Type"}, inplace=True)
output
```

```python
output.to_excel(pwd + "/Final_Output.xlsx", index=False)
```
